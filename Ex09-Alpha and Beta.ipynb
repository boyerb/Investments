# Example Chapter 9: Estimating Alpha and Beta
In this example, we estimate the alphas and betas for decile portfolios posted in the Ken French Data Library.

### Imports and Setup

```python
# Import packages
%pip install -q simple-finance
import simple_finance as sf
import requests
import pandas as pd
import numpy as np

# Establish display options when printing DataFrames
pd.set_option('display.max_columns', None)   # Show all columns without truncation
pd.set_option('display.width', 1000)   # Set the display width so output stays on one line
pd.set_option("display.max_rows", 20) # Force truncation if DataFrame has more than 20 rows
```

### Load in Data Using the Python Function
See Ex07-Ken_French.ipynb for a complete describption of the function in the course GiHub repo.


```python
# Get returns on short-term reveral deciles
df=sf.get_ff_strategies('shorttermreversal', details=True)
```

### Estaimte Alpha and Beta

```python
# Estimate alpha and beta of decile 1 relative to market portfolio.
df['dec1_excess'] = df['Dec 1'] - df['rf']

results = sf.run_ols(df['mkt-rf'], df['dec1_excess'])

print(results)
```

```python
# Estimate alpha and beta of decile 10 relative to market portfolio.
df['dec10_excess'] = df['Dec 10'] - df['rf']

results = sf.run_ols(df['mkt-rf'], df['dec10_excess'])

print(results)
```
