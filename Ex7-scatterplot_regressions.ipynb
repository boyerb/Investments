# Example Chapter 7: Regressions and Scatterplots
In this notebook, we run a regression on a scatterplot to estimate the conditional expected return given the return for the portfolio.

### Imports and Setup

```python
# Import necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from an Excel file
url = 'https://github.com/boyerb/Investments/raw/master/Examples_3.0.xlsx'
# Specify the columns to read and sheet name
columns_to_read = ['rA', 'rp']
# Load data from the specified sheet, skipping the first row and using the second row as headers
data = pd.read_excel(url, sheet_name='7-Reg1', usecols=columns_to_read, engine='openpyxl', header=1)
```

### Calculate Expected Return and Plot Regression Line
To estimate the conditional expected return for Autoliv given the return on the portfolio, we first import the data from an Excel sheet, ensuring the data is cleaned by skipping the first row of labels. The data is then stored in a DataFrame for organization. The expected return for Autoliv is calculated using the regression equation \(y = -0.003 + 1.10x\), where the portfolio return serves as the independent variable. Residuals are computed by subtracting the expected return from the actual return of Autoliv. To visually illustrate the relationship between the returns, a scatter plot is generated where each point represents a month's returns, and the regression line displays the linear relationship between the portfolio and Autoliv returns. Finally, the DataFrame's first few rows are printed to verify the accuracy of the calculations and data import process.

```python
# Calculate the conditional expected return for rA using the regression line
data['Expected_rA'] = -0.003 + 1.10 * data['rp']

# Calculate the residuals
data['Residual'] = data['rA'] - data['Expected_rA']

# Plot the scatter plot and regression line
plt.figure(figsize=(10, 6))
plt.scatter(data['rp'], data['rA'], color='blue', label='Monthly Returns')
plt.plot(data['rp'], data['Expected_rA'], color='red', label='Regression Line (y = -0.003 + 1.10x)')
plt.xlabel('Portfolio Returns (rp)')
plt.ylabel('Autoliv Returns (rA)')
plt.title('Scatter Plot and Regression Line for Monthly Returns on Autoliv and Portfolio')
plt.axhline(0, color='black', linewidth=1.5)
plt.axvline(0, color='black', linewidth=1.5)
plt.legend()
plt.grid(True)
plt.show()

# Print the first few rows of the DataFrame to check the calculations
print(data.head())
```
