# Example Chater 9: Systematic Risk and Expected Returns
In this notebook we create a plot of expected returns as a function of systematic risk

### Imports and Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define Betas and Characteristics of the Tangent Portfolio

```python
# create an array of betas from -1.5 to 1.5 by increments of 0.01
beta = np.arange(-1.5, 1.5, 0.01)

# annualized volatility of the tangent portfolio
sig_T = 0.20

# annualized risk premium of the tangent portfolio
rp_T = 0.05

# systematic risk for each beta
systematic_risk = (sig_T * beta) ** 2

# risk premium for each beta
risk_premium = rp_T * beta
```

## Create the Plot

```python
# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(systematic_risk, risk_premium, linewidth=3.5)

# Add labels and title
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Systematic Risk', fontsize=16)
plt.ylabel('Risk Premium', fontsize=16)
plt.axhline(0, color='gray', linestyle='--', linewidth=1.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=1.5)

# Show the plot
plt.show()
```
