# Example Chapter 7: Regressions and Residuals
In this notebook, we estimate regression parameters (slope and intercept) and analyze residuals.

### Imports and Setup

```python
# Import necessary packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Estimate Regression Parameters and Analyze Residuals
Here we estimate the regression parameters for a basketball dataset, where the goal is to understand the relationship between the top scorer's average points per game (PPG) and the team's win percentage (%WINS). We begin by importing the necessary packages and creating a DataFrame containing the data. It calculates the mean values of the independent variable (PPG) and dependent variable (%WINS). The covariance and variance are computed next, allowing for the calculation of the slope (b) and intercept (a) using regression formulas. The expected win percentage and residuals are calculated and stored in the DataFrame. Finally, a scatter plot is generated to visualize the data alongside the regression line, and the slope and intercept values are printed for reference.

```python
# Data provided: Season, %Wins, and Average Points per Game by Top Scorer
data = {
    'Season': [2015, 2014, 2013, 2012, 2011],
    '%WINS': [0.488, 0.463, 0.305, 0.524, 0.545],
    'AVG_PPG': [19.7, 19.3, 16.2, 17.8, 19.2]
}

# Load data into a pandas DataFrame
df = pd.DataFrame(data)

# Calculate the mean of x (AVG_PPG) and y (%WINS)
mean_x = df['AVG_PPG'].mean()
mean_y = df['%WINS'].mean()

# Calculate the covariance and variance
covariance = ((df['AVG_PPG'] - mean_x) * (df['%WINS'] - mean_y)).sum()
variance_x = ((df['AVG_PPG'] - mean_x) ** 2).sum()

# Calculate the slope (b) and intercept (a)
b = covariance / variance_x
a = mean_y - b * mean_x

# Calculate the expected %WINS and residuals
df['Expected_%WINS'] = a + b * df['AVG_PPG']
df['Residual'] = df['%WINS'] - df['Expected_%WINS']

# Plot the scatter plot and regression line with adjusted scale
plt.figure(figsize=(10, 6))
plt.scatter(df['AVG_PPG'], df['%WINS'], color='blue', label='Actual Data')
plt.plot(df['AVG_PPG'], df['Expected_%WINS'], color='red', label=f'Regression Line: y = {a:.3f} + {b:.3f}x')
plt.axhline(0, color='black', linewidth=1.5)
plt.axvline(0, color='black', linewidth=1.5)
plt.xlim(min(df['AVG_PPG']) - 1, max(df['AVG_PPG']) + 1)
plt.ylim(min(df['%WINS']) - 0.05, max(df['%WINS']) + 0.05)
plt.xlabel('Average PPG by Top Scorer')
plt.ylabel('% Wins')
plt.title('Regression Analysis: % Wins vs. Top Scorer PPG')
plt.legend()
plt.grid(True)
plt.show()

# Print the results
print("Intercept (a):", a)
print("Slope (b):", b)
print("Data with Calculations:")
print(df)
```
